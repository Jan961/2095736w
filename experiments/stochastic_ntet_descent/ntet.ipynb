```python
from experiments.cube.Cube import Cube
from hdimvis.algorithms.stochastic_ntet_algo.SNaD import SNaD
from hdimvis.create_low_d_layout.LowDLayoutCreation import LowDLayoutCreation
from hdimvis.metrics.stress.stress import vectorised_stress
from hdimvis.metrics.distance_measures.euclidian_and_manhattan import euclidean
```

```python
cube = Cube(num_points=100, side=30, angle=0.4)
dataset= cube.get_sample_dataset(3000)
```

```python
ntets= [2, 3, 4, 6, 8, 10, 12, 20, 50]

for i in ntets:
    snad = SNaD(dataset=dataset, ntet_size=i)
    layout_snad = LowDLayoutCreation().create_layout(snad, no_iters=1000, optional_metric_collection=None)
    stress = vectorised_stress(dataset.data, layout_snad.get_final_positions(), euclidean)
    cube.plot_2d(layout_snad, title=f"Snad embedding, n = {i}, stress: {stress:.3f},"
                                    f" average ntet stress: {snad.get_average_quartet_stress()}")

```

```python
snad = SNaD(dataset=dataset, ntet_size=2)
layout_snad_2 = LowDLayoutCreation().create_layout(snad, no_iters=100, optional_metric_collection=None)
stress = vectorised_stress(dataset.data, layout_snad_2.get_final_positions(), euclidean)
cube.plot_2d(layout_snad_2, title=f"Snad embedding, n = {2}, stress: {stress:.3f},"
                                f" average ntet stress: {snad.get_average_quartet_stress()}")
```

```python
from experiments.utils.layout_shannon_entropy import calculate_entropy
from experiments.utils.LayoutHistogram import LayoutHistogram

histogram_96 = LayoutHistogram(layout=layout_snad_2, pdf=True)
histogram_96.show_histogram()
print(f"Shannon entropy: {calculate_entropy(layout_snad_2,)}")
```
